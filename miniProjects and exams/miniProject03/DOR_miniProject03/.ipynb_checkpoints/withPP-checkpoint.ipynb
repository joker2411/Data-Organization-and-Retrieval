```python
import datetime
```

```python
import glob

all_files_path = glob.glob('20_newsgroups/rec.sport.hockey/*')
all_files_path
```

```python
from nltk.tokenize import word_tokenize
# from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
# tokenizer = RegexpTokenizer(r"\w+")
a = set(stopwords.words("english"))
```

```python
def remove_punc(string):
    punc = '''=!()+-[]{};:'"\, <>./?@#$%^&*_~'''
    for ele in string:  
        if ele in punc:  
            string = string.replace(ele, "") 
    return string
```

```python
doc_number = 0
IDF_dict = {}
vali = 0
lemma_words = []
punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
for current_path in all_files_path:
    doc_number += 1
    with open(current_path) as f:
        try:
            data_file = f.read().strip().lower()
        except:
            continue
        lemma_words.clear()
        tokens = word_tokenize(data_file) #tokenizer.tokenize(data_file) 
#         print("Length of all tokens : " + str(len(tokens)))
#         print(tokens)
        punct_removed_tokens = [remove_punc(i) for i in tokens]
        while('' in punct_removed_tokens) :
            punct_removed_tokens.remove('')
#         print(punct_removed_tokens)
        
        after_stopwords = [x for x in punct_removed_tokens if x not in a]
#         print("Length of all tokens after stopwords removal : " + str(len(after_stopwords)))
#         print(after_stopwords)
        
        lemma_words = [lemmatizer.lemmatize(word) for word in after_stopwords]
#         print(lemma_words)
        
        vali += 1
        
        for word in lemma_words:
            if word in IDF_dict.keys():
                if IDF_dict[word][-1] != doc_number:
                    IDF_dict[word].append(doc_number)
            else:
                IDF_dict[word] = [doc_number]
#         break
        
print(vali)
IDF_dict
```

```python
len(IDF_dict)
```

```python
sorted_IDF_dict = {k : v for k, v in sorted(IDF_dict.items())}

sorted_IDF_dict
```

```python
def or_postings(posting1, posting2):
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):
        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            result.append(posting2[p2])
            p2 += 1
        else:
            result.append(posting1[p1])
            p1 += 1
    while p1 < len(posting1):
        result.append(posting1[p1])
        p1 += 1
    while p2 < len(posting2):
        result.append(posting2[p2])
        p2 += 1
    return result
```

```python
def and_postings(posting1, posting2):
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):
        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            p2 += 1
        else:
            p1 += 1
    return result
```

```python
def posting_list_in_sorted_IDF_dict(term):
    new_term = lemmatizer.lemmatize(term.lower())
    if new_term in sorted_IDF_dict:
        return sorted_IDF_dict[new_term]
    else:
        new_list = []
        return new_list
```

```python
term1 = 'Konstantinov'
term2 = 'Montreal'
term3 = 'Ottawa'
term4 = 'canadian'
```

```python
list1 = posting_list_in_sorted_IDF_dict(term1)
list2 = posting_list_in_sorted_IDF_dict(term2)
list3 = posting_list_in_sorted_IDF_dict(term3)
list4 = posting_list_in_sorted_IDF_dict(term4)
print(list1)
print(list2)
print(list3)
print(list4)
```

```python
result_list = or_postings(and_postings(list1,list2),and_postings(list3,list4))
print(result_list)
```

```python
result_list = and_postings(or_postings(list1,list2),or_postings(list3,list4))
print(result_list)
```

# INTERESTING OUTCOME

Time consumed in query calculation without preprocessing is greater than time consumed in query calculation with preprocessing. This could be mainly because of changing every string into lower case and then converting it into it's lemma will result into increased number of postings of a string (if it's not dropped from the postings list).
Therefore, while calculating the query results, algorithm with preprocessing is taking more time in this example's case.

Note : After performing the experiment repetetively multiple times, on few occassions the value of query processing in case of algorithm with pre-processing is coming out to be fewer for same example as above. This could be mainly because system is storing data in cache to make calculations faster.
