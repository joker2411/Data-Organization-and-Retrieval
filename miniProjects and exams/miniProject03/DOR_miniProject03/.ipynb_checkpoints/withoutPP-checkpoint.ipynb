```python
import datetime
```

```python
import glob

all_files_path = glob.glob('20_newsgroups/rec.sport.hockey/*')
all_files_path
```

```python
doc_number = 0
IDF_dict = {}
vali = 0
words = []
for current_path in all_files_path:
    doc_number += 1
    with open(current_path, "r") as f:
        try:
            data_file = f.read().strip().split('\n')
        except:
            continue
#         print(data_file)
        vali += 1
        words.clear()
        for line in data_file:
            words = words + line.split(' ')
        while('' in words) :
            words.remove('')
        for word in words:
            if word in IDF_dict.keys():
                if IDF_dict[word][-1] != doc_number:
                    IDF_dict[word].append(doc_number)
            else:
                IDF_dict[word] = [doc_number]
print(vali)
IDF_dict
```

```python
len(IDF_dict)
```

```python
sorted_IDF_dict = {k : v for k, v in sorted(IDF_dict.items())}

sorted_IDF_dict
```

```python
def or_postings(posting1, posting2):
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):
        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            result.append(posting2[p2])
            p2 += 1
        else:
            result.append(posting1[p1])
            p1 += 1
    while p1 < len(posting1):
        result.append(posting1[p1])
        p1 += 1
    while p2 < len(posting2):
        result.append(posting2[p2])
        p2 += 1
    return result
```

```python
def and_postings(posting1, posting2):
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):
        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            p2 += 1
        else:
            p1 += 1
    return result
```

```python
def posting_list_in_sorted_IDF_dict(term):
    if term in sorted_IDF_dict:
        return sorted_IDF_dict[term]
    else:
        new_list = []
        return new_list
```

```python
term1 = 'Konstantinov'
term2 = 'Montreal'
term3 = 'Ottawa'
term4 = 'canadian'
```

```python
list1 = posting_list_in_sorted_IDF_dict(term1)
list2 = posting_list_in_sorted_IDF_dict(term2)
list3 = posting_list_in_sorted_IDF_dict(term3)
list4 = posting_list_in_sorted_IDF_dict(term4)
print(list1)
print(list2)
print(list3)
print(list4)
```

```python
or_postings(and_postings(list1,list2),and_postings(list3,list4))
```
