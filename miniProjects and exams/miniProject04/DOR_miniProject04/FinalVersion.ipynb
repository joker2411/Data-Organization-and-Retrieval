```python
import datetime
import glob
```

```python
all_files_path = dict()
all_files = 0
```

```python
def get_all_files():
    global all_files_path, all_files
    document = glob.glob('20_newsgroups/rec.sport.hockey/*')
    all_files = len(document)
    all_files_path = dict(zip(range(all_files), document))
```

```python
get_all_files()
all_files_path
```

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words("english"))
STOPWORDS
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
def create_tokens(doc):
    terms = word_tokenize(doc)
    terms = [term.lower() for term in terms if term not in STOPWORDS]
    return terms
```

```python
def remove_punctuation(lis):
    punc = '''=!()+-[]{};1234567890`:'"\, <>./?@#$%^&*_~'''
    new_lis = []
    for string in lis:
        for ele in string:  
            if ele in punc:  
                string = string.replace(ele, "") 
        new_lis.append(string)
    return new_lis
```

```python
def remove_blanks(lis):
    while('' in lis):
            lis.remove('')
    return lis
```

```python
def create_lemma(lis):
    lis = [lemmatizer.lemmatize(word) for word in lis]
    return lis
```

```python
import math
import re
import sys
import heapq
from collections import defaultdict
from functools import reduce
```

```python
vocabulary = set()
postings = defaultdict(dict)
document_frequency = defaultdict(int)
```

```python
def find_postings():
    global vocabulary, postings
    for id in all_files_path:
        with open(all_files_path[id], "r") as f:
            try:
                data_file = f.read().strip().lower()
            except:
                continue
                
        tokens = create_tokens(data_file)
        tokens = remove_punctuation(tokens)
        tokens = remove_blanks(tokens)
        tokens = create_lemma(tokens)
        
        unique_tokens = set(tokens)
        vocabulary = vocabulary.union(unique_tokens)
        
        for term in unique_tokens:
            postings[term][id] = tokens.count(term)
            
    postings = {k : v for k, v in sorted(postings.items())}
```

```python
find_postings()
postings
```

```python
def find_document_frequencies():
    global document_frequency
    for term in vocabulary:
        document_frequency[term] = len(postings[term])
```

```python
find_document_frequencies()
document_frequency
```

```python
def term_frequency(term, id):
    if id in postings[term]:
        return postings[term][id]
    else:
        return 0.0
```

```python
def IDF_score(term):
    if term in vocabulary:
        return math.log(all_files/document_frequency[term] , 10)
    else:
        return 0.0
```

```python
def TF_score(term, id):
    if term in vocabulary:
        return math.log(1+term_frequency(term, id), 10)
    else:
        return 0.0
```

```python
def Normalize(scores):
    length = 0.0
    for pair in scores:
        length += pair[1]**2
    length = math.sqrt(length)
    new_scores = [(pair[0],pair[1]/length) for pair in scores]
    return new_scores
```

```python
def similarity(query, id):
    similarity = 0.0
#     if(length[id] == 0):
#         return 0.0Steve Summers and the Chief were on 48 Hours last night shmoozing sports.
    for term in query:
        if term in vocabulary:
            similarity += TF_score(term, id) * IDF_score(term)
    return similarity
```

```python
def get_sorted_data(scores, K):
    #it'll reduce the sorting complexity to almost linear time in case of smaller queries
    new_scores = heapq.nlargest(K, scores, key = lambda x:x[1])
    return new_scores
```

```python
def do_search():
    query_tokens = create_tokens(input("Search query >> "))
    query_tokens = remove_punctuation(query_tokens)  
    query_tokens = remove_blanks(query_tokens)
    query_tokens = create_lemma(query_tokens)
    
#     unique_tokens = set(tokens) #We can't take ony unique tokens bcoz then we'll miss out on multiple occuring terms
    
    if query_tokens == []:
        sys.exit()
    
    K = int(input("Enter K value >> "))
    
    scores = [(id, similarity(query_tokens, id)) for id in range(all_files)] #To handle same values in query.
    scores = Normalize(scores) #To Normalize all values in the range[0,1]
    scores = get_sorted_data(scores, K) #To possibly reduce the sorting complexity
    return scores
```

```python
#This code snippet is taken from Saurabh sir's code as it was printing output very neatly.

def print_scores(scores):
    """Prints scores in a tabular format with two columns like
    | Score | Document |
    --------------------
    | 0.523 | foo      |
    --------------------

    :param scores: list of (id, score)
    """

    print("-" * 42)
    print("| %s | %-30s |" % ("Score", "Document"))
    print("-" * 42)

    for (id, score) in scores:
        if score != 0.0:
            print("| %s | %-30s |" % (str(score)[:5], all_files_path[id]))

    print("-" * 42, end="\n\n")
```

```python
def main():
    get_all_files()
    find_postings()
    find_document_frequencies()
    
    while True:
        scores = do_search()
        print_scores(scores)
```

```python
main()
```
