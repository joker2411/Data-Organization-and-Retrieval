```python
all_files_path = dict()
all_files = 0
```

```python
def get_all_files():
    global all_files_path, all_files
    document = glob.glob('20_newsgroups/rec.sport.hockey/*')
    all_files = len(document)
    all_files_path = dict(zip(range(all_files), document))
```

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words("english"))
STOPWORDS
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
def create_tokens(doc):
    terms = word_tokenize(doc)
    terms = [term.lower() for term in terms if term not in STOPWORDS]
    return terms
```

```python
def remove_punctuation(lis):
    punc = '''=!()+-[]{};1234567890`:'"\, <>./?@#$%^&*_~'''
    new_lis = []
    for string in lis:
        for ele in string:  
            if ele in punc:  
                string = string.replace(ele, "") 
        new_lis.append(string)
    return new_lis
```

```python
def remove_blanks(lis):
    while('' in lis):
            lis.remove('')
    return lis
```

```python
def create_lemma(lis):
    lis = [lemmatizer.lemmatize(word) for word in lis]
    return lis
```

```python
import math
import re
import sys
from collections import defaultdict
from functools import reduce
```

```python
vocabulary = set()
postings = defaultdict(dict)
document_frequency = defaultdict(int)
length = defaultdict(float)
```

```python
def find_postings():
    global vocabulary, postings
    for id in all_files_path:
        with open(all_files_path[id], "r") as f:
            try:
                data_file = f.read().strip().lower()
            except:
                continue
#         print(data_file)
        
        tokens = create_tokens(data_file)
        
        tokens = remove_punctuation(tokens)
        
        tokens = remove_blanks(tokens)
#         print(tokens)
        
        tokens = create_lemma(tokens)
#         print(len(tokens))
        
        unique_tokens = set(tokens)
#         print(len(unique_tokens))
        
        vocabulary = vocabulary.union(unique_tokens)
#         print(vocabulary)
        
        for term in unique_tokens:
            postings[term][id] = tokens.count(term)
#         print(postings)
    postings = {k : v for k, v in sorted(postings.items())}
```

```python
def find_document_frequencies():
    global document_frequency
    for term in vocabulary:
        document_frequency[term] = len(postings[term])
```

```python
def term_frequency(term, id):
    if id in postings[term]:
        return postings[term][id]
    else:
        return 0.0
```

```python
def IDF_score(term):
    if term in vocabulary:
        return math.log(all_files/document_frequency[term] , 10)
    else:
        return 0.0
```

```python
def TF_score(term, id):
    if term in vocabulary:
        return math.log(1+term_frequency(term, id), 10)
    else:
        return 0.0
```

```python
def initialize_lengths():
    global length
    for id in all_files_path:
        l = 0
        for term in vocabulary:
            l += term_frequency(term, id) ** 2
        length[id] = math.sqrt(l)
```

```python
initialize_lengths()
```

```python
length
```

```python
def similarity(query, id):
    similarity = 0.0
    if(length[id] == 0):
        return 0.0
    for term in query:
        if term in vocabulary:
            similarity += term_frequency(term, id) * IDF_score(term)
    similarity = similarity / length[id]
    return similarity
```

```python
def do_search():
    query_tokens = create_tokens(input("Search query >> "))
    
    query_tokens = remove_punctuation(query_tokens)
        
    query_tokens = remove_blanks(query_tokens)
#         print(tokens)
        
    query_tokens = create_lemma(query_tokens)
#         print(len(tokens))
        
#     unique_tokens = set(tokens)
    

    if query_tokens == []:
        sys.exit()
    
    K = int(input("Enter K value >> "))

    scores = sorted(
        [(id, similarity(query_tokens, id)) for id in range(all_files)], key=lambda x: x[1],reverse=True,)

    return scores[0:K]

```

```python
def print_scores(scores):
    """Prints scores in a tabular format with two columns like
    | Score | Document |
    --------------------
    | 0.523 | foo      |
    --------------------

    :param scores: list of (id, score)
    """
    print("-" * 42)
    print("| %s | %-30s |" % ("Score", "Document"))
    print("-" * 42)

    for (id, score) in scores:
        if score != 0.0:
            print("| %s | %-30s |" % (str(score)[:5], all_files_path[id]))

    print("-" * 42, end="\n\n")
```

```python
def main():
    get_all_files()
    find_postings()
    find_document_frequencies()
    initialize_lengths()
    
    while True:
        scores = do_search()
        print_scores(scores)
```

```python
main()
```
